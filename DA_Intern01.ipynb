```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mcdonalds/nutrition-facts")

print("Path to dataset files:", path)
```

```python
import os

dataset_path = "/root/.cache/kagglehub/datasets/mcdonalds/nutrition-facts/versions/1"

os.listdir(dataset_path)
```

```python
import pandas as pd

file_path = dataset_path + "/menu.csv"   # update filename if different
df = pd.read_csv(file_path)

df.head()
```

```python
df.shape          # rows, columns
df.info()         # data types
df.isnull().sum() # missing values
```

```python
df = df.drop_duplicates()
df = df.fillna(0)
```

```python
df.describe()
```

```python
df["Calories"].mean()
df["Calories"].median()
df["Calories"].mode()
df["Calories"].std()
```

Menu Item = Retail Product

Category = Product Category

Calories / Price proxy = Sales value indicator

Item frequency = Customer demand

```python
category_counts = df["Category"].value_counts()
category_counts
```

```python
import matplotlib.pyplot as plt

category_counts.plot(kind="bar", figsize=(8,5))
plt.title("Number of Products per Category")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.show()
```

```python
df["Calories"].describe()
```

```python
mean_cal = df["Calories"].mean()
median_cal = df["Calories"].median()
mode_cal = df["Calories"].mode()[0]
std_cal = df["Calories"].std()

mean_cal, median_cal, mode_cal, std_cal
```

```python
top_items = df.groupby("Category")["Calories"].mean().sort_values(ascending=False)
top_items
```

```python
import matplotlib.pyplot as plt

top_items = df.groupby("Category")["Calories"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))   # increase width
plt.plot(top_items.index, top_items.values, marker="o")

plt.title("Average Calories by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Calories")

plt.xticks(rotation=45, ha="right")  # rotate labels
plt.tight_layout()                   # auto adjust spacing
plt.show()
```

```python
import seaborn as sns

numeric_cols = df.select_dtypes(include="number")
corr = numeric_cols.corr()

plt.figure(figsize=(10,7))   # bigger figure
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    annot_kws={"size": 8}     # smaller text
)

plt.title("Correlation Heatmap of Nutritional Features")
plt.tight_layout()
plt.show()
```

```python
df["Item_Index"] = range(1, len(df)+1)

plt.plot(df["Item_Index"], df["Calories"])
plt.title("Simulated Retail Trend")
plt.xlabel("Product Index")
plt.ylabel("Calories")
plt.show()
```
